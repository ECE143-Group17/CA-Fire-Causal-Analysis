```python
import pandas as pd
from altair import Chart, load_dataset
import altair as alt
alt.renderers.enable('notebook')


dfile='2015_Redbook_FINAL.xlsx'
j=6

while j in range(6,9):      
    df = pd.read_excel(dfile, sheetname='Table %d'%j,skiprows=[0,1])
    chart = Chart(df) #input should be a dataframe
  
    l1+=chart.mark_point().encode(x='Start',
                              y='Total',
                             color='DPA')
    j+=1
    
  
l1

```
