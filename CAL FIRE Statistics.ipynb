# Cause

```python
cause = ['Arson','Campfire','Debris Burning','Elec. Power','Equip. Use','Lightening','Miscellaneous',
        'Play with Fire','Railroad','Smoking','Undetermined','Vehicle']
```

```python
df_16 = pd.read_excel("2016_Redbook_FINAL.xlsx",sheet_name='Table 12',skiprows=[0])
```

```python
df_16.index = df_16[df_16.columns[0]].tolist()
```

```python
df16_north = df_16.loc['TOTAL'][0:1]
df16_south = df_16.loc['TOTAL'][1:2]
```

```python
def causes_total(df):
    causes_total = []
    for i in range(2,len(df.loc['STATE TOTAL'])):
        if df.loc['STATE TOTAL'][i] == df.loc['STATE TOTAL'][i]:
            causes_total.append(df.loc['STATE TOTAL'][i])
    return causes_total
```

```python
def region_total(df):
    total = []
    for i in range(2,len(df.loc['TOTAL'])):
        if df.loc['TOTAL'][i] == df.loc['TOTAL'][i]:
            total.append(df.loc['TOTAL'][i])
    return total
```

```python
causes_16 = causes_total(df_16)
```

```python
north_16 = region_total(df16_north)
south_16 = region_total(df16_south)
```

```python
colors=[ '#9999ff', '#ff9999', '#7777aa', '#2442aa', '#dd5555','#CD5555','#B9D3EE','#B22222','#7A378B','#9370DB','#388E8E','#AEEEEE']
plt.figure(figsize=(10,10))
plt.pie(causes_16,labels=cause,startangle=90,labeldistance = 1.1,colors=colors,autopct= '%.1f%%',
        wedgeprops = { 'linewidth': 1, 'edgecolor': 'black'})

plt.axis('equal')
plt.title('PERCENTAGE OF CAUSE OF FIRES 2016',fontsize=15)

plt.show()
```

```python
df_15 = pd.read_excel("2015_Redbook_FINAL.xlsx",sheet_name='Table 13',skiprows=[0])
df_14 = pd.read_excel("2014_Redbook_Complete.xlsx",sheet_name='Table 13',skiprows=[0])
```

```python
df_15.index = df_15[df_15.columns[0]].tolist()
df_14.index = df_14[df_14.columns[0]].tolist()
```

```python
causes_15 = causes_total(df_15)
causes_14 = causes_total(df_14)
```

```python
x = np.arange(len(cause))
colors = ['#3A5FCD','#CD6090','#C6E2FF']
fig,ax = plt.subplots(figsize=(20,10))
b1 = ax.bar(x-0.25,causes_14,0.25,color=colors[0])
b2 = ax.bar(x,causes_15,0.25,color=colors[1],tick_label=cause)
b3 = ax.bar(x+0.25,causes_16,0.25,color=colors[2])

plt.xlabel('CAUSES',fontsize=20)
plt.ylabel('NUMBER OF FIRES',fontsize=20)
plt.xticks(fontsize=13)
plt.yticks(fontsize=15)
plt.title('FIRES BY CAUSE 2014,2015,2016',fontsize=20)
plt.legend((b1,b2,b3),('2014','2015','2016'),fontsize=20)
plt.show()
```
